```python
#Stored the data to DBs by class
```

```python

DB_shortT=pd.concat([taobao_shortT,asos_shortT,amazon_shortT,adidasofficial_shortT])
DB_shortT.reset_index(drop=True, inplace=True)
DB_shortT['index'] = DB_shortT.index

DB_longT=pd.concat([taobao_longT,asos_longT,amazon_longT,adidasofficial_longT])
DB_longT.reset_index(drop=True, inplace=True)
DB_longT['index'] = DB_longT.index

DB_poloT=pd.concat([taobao_poloT,asos_poloT,amazon_poloT,adidasofficial_poloT])
DB_poloT.reset_index(drop=True, inplace=True)
DB_poloT['index'] = DB_poloT.index


DB_hoodies=pd.concat([taobao_hoodies,asos_hoodies,amazon_hoodies,adidasofficial_hoodies])
DB_hoodies.reset_index(drop=True, inplace=True)
DB_hoodies['index'] = DB_hoodies.index

DB_shorts=pd.concat([taobao_shorts,asos_shorts,amazon_shorts,adidasofficial_shorts])
DB_shorts.reset_index(drop=True, inplace=True)
DB_shorts['index'] = DB_shorts.index

DB_pants=pd.concat([taobao_pants,asos_pants,amazon_pants,adidasofficial_pants])
DB_pants.reset_index(drop=True, inplace=True)
DB_pants['index'] = DB_pants.index

DB_sportbra=pd.concat([taobao_sportbra,adidasofficial_sportbra])
DB_sportbra.reset_index(drop=True, inplace=True)
DB_sportbra['index'] = DB_sportbra.index

DB_jacket=pd.concat([asos_jacket,amazon_jacket,adidasofficial_jacket])
DB_jacket.reset_index(drop=True, inplace=True)
DB_jacket['index'] = DB_jacket.index
```
