```python
def get_feature_vector(img):
 image = cv2.resize(img, (224, 224),3)
 feature_vector = basemodel.predict(image.reshape( 1, 224, 224,3))
 return feature_vector
```

```python
def calculate_similarity(vector1, vector2):
 return 1- scipy.spatial.distance.cosine(vector1, vector2)
```

```python
basemodel=load_model('vgg19.h5')
```

```python
#User upload part
original = cv2.imread("image_Polo_1397.jpg")
org2=cv2.resize(original, (800, 800)) 
#Fill the edge
org2[0:80,]=0
org2[720:800,]=0
org2[:,0:80]=0
org2[:,720:800]=0
f1 = get_feature_vector(org2)
```

```python
#load class from yolo
yoloclass=('shortT','pants','adidas')
```

```python
#load class from yolo
yoloclass=('shortT','pants','adidas')
if 'Adidas' in yoloclass:
    print('We find Adidas is in your upload image. Please check our recommendations.')
else: print('Please check our recommendations.')
```

```python
v_shortT=[]
v_longT=[]
v_poloT=[]
v_hoodies=[]
v_shorts=[]
v_pants=[]
v_sportbra=[]
v_jacket=[]
```

```python
# 2) loop correspondong class and use VGG19 and distance.cosine to find the image similarity
for i in yoloclass:
    if i=='Short-T':
        for i in range(1842):
            v_shortT.append(calculate_similarity(f1[0], np.loadtxt('./vectors/shortT_f2_'+str(i)+'.txt', delimiter=','))*100)
    elif i=='Long-T':
        for i in range(529):
            v_longT.append(calculate_similarity(f1[0], np.loadtxt('./vectors/longT_f2_'+str(i)+'.txt', delimiter=','))*100)
    elif i=='Polo-T':
        for i in range(385):
            v_poloT.append(calculate_similarity(f1[0], np.loadtxt('./vectors/poloT_f2_'+str(i)+'.txt', delimiter=','))*100)
    elif i=='Hoodie':
        for i in range(1287):
            v_hoodies.append(calculate_similarity(f1[0], np.loadtxt('./vectors/hoodies_f2_'+str(i)+'.txt', delimiter=','))*100)
    elif i=='Shorts':
        for i in range(1359):
            v_shorts.append(calculate_similarity(f1[0], np.loadtxt('./vectors/shorts_f2_'+str(i)+'.txt', delimiter=','))*100)
    elif i=='Pants':
        for i in range(1450):
            v_pants.append(calculate_similarity(f1[0], np.loadtxt('./vectors/pants_f2_'+str(i)+'.txt', delimiter=','))*100)
    elif i=='SportBra':
        for i in range(54):
            v_sportbra.append(calculate_similarity(f1[0], np.loadtxt('./vectors/sportbra_f2_'+str(i)+'.txt', delimiter=','))*100)
    elif i=='Jacket':
        for i in range(len(614)):
            v_jacket.append(calculate_similarity(f1[0], np.loadtxt('./vectors/jacket_f2_'+str(i)+'.txt', delimiter=','))*100)
```
