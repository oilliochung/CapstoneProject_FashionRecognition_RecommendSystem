```python
import pandas as pd
import numpy as np
#### download image in one click ####
import urllib.request
```

```python
# procedure that retrieves to URL and save the image as JPG file to define directory
def url_to_jpg(i, url, filepath):
    # i = number of image
    # url = a URL address of a given image
    # filepath = where to save the final image
    
    filename = 'image_shortT_{}.jpg'.format(i)
    fullpath = '{}{}'.format(filepath, filename)
    urllib.request.urlretrieve(url, fullpath)

    print('{} saved.'.format(filename))

    return None
# set constants
filename = 'C:/Users/oillio/Desktop/FinalProject/1_Gather_data/imgurlcsv/shortT.csv'
filepath = 'C:/Users/oillio/Desktop/FinalProject/1_Gather_data/images/shortT/'
# read list of urls as dandas dataframe
urls = pd.read_csv(filename)
# save images to the directory by iterating through the list
for i, url in urls.values:
    url_to_jpg(i, url, filepath)
```
