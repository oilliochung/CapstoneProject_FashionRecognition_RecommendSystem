```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
import regex as re
import numpy as np
import requests
import string
import time 
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#short tee #Done
domain1 = 'https://world.taobao.com/product/Adidas%E7%9F%AD%E8%A2%96_' #short tee Link
domain2 = '.htm?spm=a21wu.10013406-cat-tw.0.0.1ab725fb0ePYoS' #short tee Link
URL = 'https://world.taobao.com/product/Adidas%E7%9F%AD%E8%A2%96.htm?spm=a21wu.10013406-cat-tw.0.0.1ab725fb0ePYoS' #short tee Link
driver = webdriver.Chrome(ChromeDriverManager().install())
#driver.get(URL)
#time.sleep(120)
#subhtml = driver.page_source
#soup = BeautifulSoup(subhtml, "html.parser")

df = pd.DataFrame() #short tee
for i in range(1,30): #page 1-30
        nextlink = domain1 + str(i) +domain2
        driver.get(nextlink)
        
        # Smooth Scorlling
        total_height = int(driver.execute_script("return document.body.scrollHeight"))
        for i in range(1, total_height, 5):
            driver.execute_script("window.scrollTo(0, {});".format(i))
            time.sleep(0.01)
        
        subhtml = driver.page_source
        soup = BeautifulSoup(subhtml, "html.parser")
        
        for i in soup.find_all(class_='item-box'): #scrape 1-page data
            img='https:'+i.img['src']
            price=i.find(class_="price").text
            sales=i.find(class_="sales").text
            desc=i.find(class_="item-desc").text
            url=i.find('a')['href']
            df = df.append({'type':'short tee', 'img':img, 'price':price, 'sales':sales, 'desc':desc, 'url':url}, ignore_index = True)
```
